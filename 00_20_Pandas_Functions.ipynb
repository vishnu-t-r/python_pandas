#### 15) pd.pivot_table()

```python
#create a pivot from a dataframe
#df.info()

#total sales by segment
sales_segment = pd.pivot_table(df,
                               index='Segment',
                              values='Sales',
                              aggfunc='sum')
sales_segment
```

```python
#total sales by segment and shipmode
shp_seg_sales = pd.pivot_table(df,
                                index=['Segment','Ship Mode'],
                                values='Sales',
                                aggfunc='sum')
shp_seg_sales
```

```python
#total sales and profit by segment and shipmode
shp_seg_sales_pro = pd.pivot_table(df,
                                index=['Segment','Ship Mode'],
                                values=['Sales','Profit'],
                                aggfunc='sum')
shp_seg_sales_pro
```

#### 18)df.merge()

```python
#similar to joins
```

```python
#create first datarframe 
m1 = pd.DataFrame({'key':['A','B','C','D'],
                   'value':[1,2,3,4]})
m1
```

```python
#create second dataframe
m2 = pd.DataFrame({'key':['B','D','E','F'],
                  'value':[5,6,7,8]})
m2
```

```python
#merge the dataframe m1 and m2 based on key column
m3 = m1.merge(m2,on='key')
m3
```

```python
#using the 'how' parameter type of join can be specified (left/right/outer)
m4 = m1.merge(m2,how='right',on='key')
m4

#using multiple columns with the on parameter
#m5 = m1.merge(m2,on=['key1','key2'])

#specify left join and right join on seperate columns
#m5 = m1.merge(m2,left_on='key1',right_on='key2')
```
